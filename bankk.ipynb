```python
import random

class Account:
    # 클래스 변수
    account_count = 0

    # 생성자 메서드
    def __init__(self, name, balance):
        self.bank = "Hungry Bank"
        self.name = name
        self.balance = balance             # 초기 잔액
        self.account_number = self.make_account()  # 계좌번호 생성

        self.deposit_count = 0             # 입금 횟수
        self.deposit_log = []              # 입금 내역
        self.withdraw_log = []             # 출금 내역

        Account.account_count += 1         # 계좌 개설 시 클래스 변수 +1

    # 계좌번호 생성 메서드
    def make_account(self):
        num1 = str(random.randint(0, 999)).zfill(3)     # 3자리
        num2 = str(random.randint(0, 99)).zfill(2)      # 2자리
        num3 = str(random.randint(0, 999999)).zfill(6)  # 6자리
        return f"{num1}-{num2}-{num3}"

    # 클래스 변수 출력 메서드
    @classmethod
    def get_account_num(cls):
        print(f"현재 생성된 계좌 수: {cls.account_count}개")

    # Q4. 입금 메서드
    def deposit(self, amount):
        if amount < 1:
            print(" 입금 금액은 1원 이상이어야 합니다.")
            return

        self.balance += amount
        self.deposit_count += 1
        self.deposit_log.append(amount)   # 입금 내역 기록

        print(f" {format(amount, ',')}원이 입금되었습니다. 현재 잔액: {format(self.balance, ',')}원")

        # 5회마다 1% 이자 지급
        if self.deposit_count % 5 == 0:
            interest = int(self.balance * 0.01)
            self.balance += interest
            print(f"입금 {self.deposit_count}회 달성! 이자 {format(interest, ',')}원이 지급되었습니다.")
            print(f" 이자 지급 후 잔액: {format(self.balance, ',')}원")

    # Q5. 출금 메서드
    def withdraw(self, amount):
        if amount > self.balance:
            print(" 출금 금액이 잔액보다 많습니다.")
        elif amount < 1:
            print("출금 금액은 1원 이상이어야 합니다.")
        else:
            self.balance -= amount
            self.withdraw_log.append(amount)  # 출금 내역 기록
            print(f"{format(amount, ',')}원이 출금되었습니다. 현재 잔액: {format(self.balance, ',')}원")

    # Q6. 정보 출력 메서드
    def display_info(self):
        print(f"(은행이름: {self.bank}, 예금주: {self.name}, 계좌번호: {self.account_number}, 잔고: {format(self.balance, ',')}원)")

    # Q10. 입금 내역 출력 메서드
    def deposit_history(self):
        print(f" [입금 내역] 예금주: {self.name}")
        if not self.deposit_log:
            print("  - 입금 내역이 없습니다.")
        else:
            for i, amount in enumerate(self.deposit_log, start=1):
                print(f"  {i}. {format(amount, ',')}원")

    # Q10. 출금 내역 출력 메서드
    def withdrawal_history(self):
        print(f"[출금 내역] 예금주: {self.name}")
        if not self.withdraw_log:
            print("  - 출금 내역이 없습니다.")
        else:
            for i, amount in enumerate(self.withdraw_log, start=1):
                print(f"  {i}. {format(amount, ',')}원")


# 계좌 3개 이상 생성
acc1 = Account("조아영", 500000)
acc2 = Account("윤지성", 1500000)
acc3 = Account("헝그리", 900000)
acc4 = Account("피자", 2000000)

# 리스트에 저장
accounts = [acc1, acc2, acc3, acc4]

print("잔고 1,000,000원 이상 고객 정보")
for acc in accounts:
    if acc.balance >= 1_000_000:
        acc.display_info()

# 몇 번 입금 / 출금
acc1.deposit(100000)
acc1.deposit(50000)
acc1.withdraw(30000)

# 내역 출력
acc1.deposit_history()
acc1.withdrawal_history()


```
