```python
class Square:
    def __init__(self):
        self.square = int(input('넓이를 구할 도형의 번호를 입력하세요.\n1. 직사각형  2. 평행사변형  3. 사다리꼴\n>>> '))

        if self.square == 1:
            print(self.rect())
        elif self.square == 2:
            print(self.par())
        elif self.square == 3:
            print(self.trape())
        else:
            print('1, 2, 3 중에서 다시 입력해주세요.')

    def rect(self):
        width, vertical = map(int, input('가로, 세로를 입력하세요. 예시: 가로,세로\n>>> ').split(','))
        area = width * vertical
        return f'직사각형의 넓이는 : {area}'

    def par(self):
        base, height = map(int, input('밑변, 높이를 입력하세요. 예시: 밑변,높이\n>>> ').split(','))
        area = base * height
        return f'평행사변형의 넓이는 : {area}'

    def trape(self):
        bottom, top, height = map(int, input('밑변, 윗변, 높이를 입력하세요. 예시: 밑변,윗변,높이\n>>> ').split(','))
        area = (bottom + top) * height / 2
        return f'사다리꼴의 넓이는 : {area}'

a = Square()
```
